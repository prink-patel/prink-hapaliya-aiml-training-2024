```python
import cv2
import threading
```

```python
img = cv2.imread('img.jpg')
```

```python
height, width,ch = img.shape
half_height,half_width=height//2,width//2
```

```python
first_img = img[:half_height,:half_width]
sec_img=img[half_height:,:half_width]
third_img=img[:half_height,half_width:]
forth_img=img[half_height:,half_width:]
```

```python
def gray_con(img):
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
    return gray_image

def conver_img(lock,img,gray_img):
    lock.acquire()
    img[:,:,0]=gray_con(gray_img)
    lock.release()
    return img
```

```python
lock=threading.Lock()

t1=threading.Thread(target=conver_img,args=(lock,first_img))
t2=threading.Thread(target=conver_img,args=(lock,sec_img))
t3=threading.Thread(target=conver_img,args=(lock,third_img))
t4=threading.Thread(target=conver_img,args=(lock,forth_img))
```

```python
cv2.imshow('img_gray.jpg', img[:,:,1]) 
cv2.waitKey(0)   
  
# Window shown waits for any key pressing event 
cv2.destroyAllWindows()
```
